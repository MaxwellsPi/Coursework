## Report: act_report

This act report includes the summary of the Data Analysis process that was conducted for the data wrangling project.

**Data**
>In this project, three dataset were used.
- The first dataset was provided as a CSV file named twitter_archive_enhanced.csv. It contains basic information about 2356 tweets and was uploaded manually into the workspace.
- The second dataset was in TSV format titles image_prediction.tsv which was hosted on udacity server and was programmatically downloaded. It contains 2075 predictions made by a neural network that was trained to classify dog breeds.
- The third dataset was in was provided as a text file in JSON format. This dataset contains information like the rewetet count, favorite count, followers count and friends count each tweet recieved for 2327 tweets in the file "tweet_json_text".

**Cleaning**
>During accessing the data, 8 quality issues and 4 tidiness issues were addressed. After addressing all quality and tidiness issues the dataset were joined by tweet_id into one master dataframe.  This was then used to gain some quick insights and visualizations.  Found in the following section.

### Insights

- Golden retrievers are the dogs most predicted by the neural network classifer.

- Samoyeds had, on average, the highest number of retweets and favorites.

- The neural net algorithm was most confident when predicting Pugs.
